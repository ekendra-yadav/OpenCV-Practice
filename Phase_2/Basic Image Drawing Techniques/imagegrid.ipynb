```python
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import numpy as np
import os
from matplotlib.ticker import MultipleLocator

# 1. PASTE YOUR COPIED PATH HERE
# Example: r'C:\Users\Admin\Desktop\OpenCV\Phase_2\Ekendra Yadav Profile.jpeg'
image_path = r'C:\Users\Admin\Desktop\OpenCV\Phase_2\Ekendra Yadav Profile.jpeg' 

# 2. Check if the file exists before processing to avoid crashes
if not os.path.exists(image_path):
    print(f"❌ Error: File not found at: {image_path}")
    print(f"Current folder Python is looking in: {os.getcwd()}")
else:
    # 3. Setup Plotting Style
    sns.set_theme(style="white") 

    # 4. Load the Image
    img = Image.open(image_path)
    img_array = np.array(img)

    # 5. Define Coordinate System (0-100 Scale)
    width, height = img.size
    aspect_ratio = height / width
    x_max = 100
    y_max = x_max * aspect_ratio

    # 6. Create the Interface
    fig, ax = plt.subplots(figsize=(12, 10))

    # Display Image (origin='upper' keeps it upright with custom extents)
    ax.imshow(img_array, extent=[0, x_max, 0, y_max], origin='upper', alpha=0.9)

    # 7. Configure Graph Paper (Major & Minor Grids)
    ax.xaxis.set_major_locator(MultipleLocator(10))
    ax.yaxis.set_major_locator(MultipleLocator(10))
    ax.xaxis.set_minor_locator(MultipleLocator(2))
    ax.yaxis.set_minor_locator(MultipleLocator(2))

    # Style lines (Blue graph paper look)
    ax.grid(which='major', color='#2980b9', linestyle='-', linewidth=1.0, alpha=0.6)
    ax.grid(which='minor', color='#3498db', linestyle=':', linewidth=0.5, alpha=0.3)

    # 8. Add Central XY Axes
    ax.axhline(y_max/2, color='black', linewidth=1.5, alpha=0.8) 
    ax.axvline(x_max/2, color='black', linewidth=1.5, alpha=0.8) 

    # 9. Final Labels
    ax.set_title("Interactive Image Graph Overlay", fontsize=15, pad=20)
    ax.set_xlabel("X-Axis", fontsize=12)
    ax.set_ylabel("Y-Axis", fontsize=12)
    ax.set_xlim(0, x_max)
    ax.set_ylim(0, y_max)

    plt.tight_layout()
    plt.show()
```
